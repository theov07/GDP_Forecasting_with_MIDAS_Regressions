```python
import pandas as pd

excel_path = "DATAS/BBG_Data_V3.xlsx"  
output_csv = "DATAS/bloomberg_all_tickers.csv"

xls = pd.ExcelFile(excel_path)

all_data = []

for sheet in xls.sheet_names:
    
    df = pd.read_excel(xls, sheet_name=sheet, header=None)

    df = df.iloc[1:].copy()

    df.columns = ["date", "price"]

    df["date"] = pd.to_datetime(df["date"], dayfirst=True, errors="coerce")

    df["ticker"] = sheet

    all_data.append(df[["date", "ticker", "price"]])


df_all = pd.concat(all_data, ignore_index=True)

df_all = df_all.sort_values(["date", "ticker"])

df_all.to_csv(output_csv, index=False, encoding="utf-8")

print(f"CSV créé : {output_csv}")
print(df_all.head())
```
